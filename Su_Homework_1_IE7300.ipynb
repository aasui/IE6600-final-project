## 1. Normal Distribution

### 1. Generate a normal distribution(mean = 6, standard deviation =2.5), for three different sample sizes 150, 1500, 150,000. Design a python class to get the following items:


#### a) Show the density curve for all the three sample sizes

```python
class ND:
    def __init__(self, mean, std_dev, sample_size):
        self.mean = mean
        self.std_dev = std_dev
        self.sample_size = sample_size

    def generate_samples(self):
        return np.random.normal(self.mean, self.std_dev, self.sample_size)

    def plot_density_curve(self):
        plt.figure(figsize=(8, 6))
        plt.title(f'Normal Distribution (Mean={self.mean}, Std Dev={self.std_dev}, Sample Size={self.sample_size})')

        # Generate random samples
        samples = self.generate_samples()

        # Create a range of values for the x-axis
        x = np.linspace(self.mean - 4 * self.std_dev, self.mean + 4 * self.std_dev, 1000)

        # Calculate the probability density function (PDF) for the normal distribution
        pdf = stats.norm.pdf(x, self.mean, self.std_dev)

        # Plot the histogram of the generated samples
        plt.hist(samples, bins=30, density=True, alpha=0.5, label='Sampled Data')

        # Plot the probability density curve
        plt.plot(x, pdf, 'r', lw=2, label='PDF')
        plt.legend()
        plt.grid(True)
        plt.xlabel('Value')
        plt.ylabel('Probability Density')
        plt.show()

# Create instances of the NormalDistribution class for different sample sizes
sample_sizes = [150, 1500, 150000]
mean = 6
std_dev = 2.5

for size in sample_sizes:
    distribution = ND(mean, std_dev, size)
    distribution.plot_density_curve()
```

This code defines a "NormalDistribution" class that can generate samples from a normal distribution with specified parameters and plot the density curve for those samples. It then creates instances of this class for the three different sample sizes and plots the density curves for each size.

#### c) Convert the simulated data into a standard normal distribution, and plot for all three different sample sizes.

```python
class ND:
    def __init__(self, mean, std_dev, sample_size):
        self.mean = mean
        self.std_dev = std_dev
        self.sample_size = sample_size

    def generate_samples(self):
        return np.random.normal(self.mean, self.std_dev, self.sample_size)

    def plot_density_curve(self, title):
        plt.figure(figsize=(8, 6))
        plt.title(title)

        # Generate random samples
        samples = self.generate_samples()

        # Convert samples to the standard normal distribution
        z_scores = (samples - self.mean) / self.std_dev

        # Create a range of values for the x-axis
        x = np.linspace(-4, 4, 1000)

        # Calculate the probability density function (PDF) for the standard normal distribution
        pdf = stats.norm.pdf(x, 0, 1)

        # Plot the histogram of the converted samples
        plt.hist(z_scores, bins=30, density=True, alpha=0.5, label='Standardized Data')

        # Plot the probability density curve for the standard normal distribution
        plt.plot(x, pdf, 'r', lw=2, label='Standard Normal PDF')
        plt.legend()
        plt.grid(True)
        plt.xlabel('Value (Standardized)')
        plt.ylabel('Probability Density')
        plt.show()

# Create instances of the NormalDistribution class for different sample sizes
sample_sizes = [150, 1500, 150000]
mean = 6
std_dev = 2.5

for size in sample_sizes:
    distribution = ND(mean, std_dev, size)
    title = f'Standardized Normal Distribution (Sample Size={size})'
    distribution.plot_density_curve(title)
```

## 2. Standard Normal Distribution

### 2. Generate a standard normal distribution for three different sample sizes 100, 5000, 50,000. Design a python class to get the following items:

#### a) Show the density curve for all the three sample sizes

```python
class SND:
    def __init__(self, sample_size):
        self.sample_size = sample_size

    def generate_samples(self):
        return np.random.standard_normal(self.sample_size)

    def plot_density_curve(self):
        plt.figure(figsize=(8, 6))
        plt.title(f'Standard Normal Distribution (Sample Size={self.sample_size})')

        # Generate random samples from the standard normal distribution
        samples = self.generate_samples()

        # Create a range of values for the x-axis
        x = np.linspace(-4, 4, 1000)

        # Calculate the probability density function (PDF) for the standard normal distribution
        pdf = stats.norm.pdf(x, 0, 1)

        # Plot the histogram of the generated samples
        plt.hist(samples, bins=30, density=True, alpha=0.5, label='Sampled Data')

        # Plot the probability density curve for the standard normal distribution
        plt.plot(x, pdf, 'r', lw=2, label='Standard Normal PDF')
        plt.legend()
        plt.grid(True)
        plt.xlabel('Value')
        plt.ylabel('Probability Density')
        plt.show()

# Create instances of the StandardNormalDistribution class for different sample sizes
sample_sizes = [100, 5000, 50000]

for size in sample_sizes:
    distribution = SND(size)
    distribution.plot_density_curve()
```

#### b) Explain the findings for all three samples

1. Sample Size = 100:


*   With a relatively small sample size (n = 100), the histogram of the sampled data may appear somewhat noisy, and it may not perfectly follow the shape of the standard normal distribution.

*   However, the histogram should still exhibit a roughly bell-shaped curve centered around 0, which is the mean of the standard normal distribution.
*   Some variation and noise may be present due to the small sample size.

2. Sample Size = 5000:


*   With a larger sample size (n = 5000), the histogram of the sampled data should closely approximate the shape of the standard normal distribution.

*   The histogram should exhibit a smoother and more symmetric bell-shaped curve centered around 0.
*   The variation and noise in the data should be significantly reduced compared to the smaller sample size.

3. Sample Size = 50,000:


*   With a very large sample size (n = 50,000), the histogram should closely match the theoretical standard normal distribution.

*   The histogram should exhibit a nearly perfect bell-shaped curve centered around 0, with minimal variation and noise.
*   The data points in the histogram should be densely packed around the mean of 0.



In summary, as the sample size increases, the sampled data more closely approximates the properties of the standard normal distribution (mean = 0, standard deviation = 1). Larger sample sizes lead to more accurate and precise representations of the theoretical distribution, while smaller sample sizes may exhibit more variation and noise. This is consistent with the Central Limit Theorem, which states that as the sample size increases, the distribution of sample means approaches a standard normal distribution.




## 3. Binomial Distribution Simulation

### 3. Binomial Distribution simulation. Generate a random number between 0 and 1. If that number is 0.5 or more, then event it as fake. Otherwise not a fake for 500, 5000, and 500,000 trails. Design a python class to get the following items:

#### a) Show the density curve for all the three sample sizes

```python
# Parameters
trials = [500, 5000, 500000]
prob = 0.5  # Probability of an event being fake

# Simulate the binomial distribution for each number of trials
results = []
for num_trials in trials:
    data = np.random.rand(num_trials) < prob
    results.append(data)

# Plot and print the density curve for each sample size
plt.figure(figsize=(12, 6))
for i, num_trials in enumerate(trials):
    plt.subplot(1, 3, i + 1)
    plt.hist(results[i], bins=[0, 0.5, 1], density=True, alpha=0.6, color='b', edgecolor='k')
    plt.title(f'Sample Size: {num_trials}')
    plt.xlabel('Fake (0.5 or more) vs. Not Fake (Less than 0.5)')
    plt.ylabel('Density')
    plt.xticks([0, 1], ['Not Fake', 'Fake'])

    # Print the data for each sample size
    fake_count = np.sum(results[i])
    not_fake_count = num_trials - fake_count
    print(f'Sample Size: {num_trials}')
    print(f'Number of Fakes: {fake_count}')
    print(f'Number of Not Fakes: {not_fake_count}')
    print('-' * 30)

plt.tight_layout()
plt.show()
```

This script will generate random numbers between 0 and 1, compare them to 0.5, and store the results for the specified number of trials. It will then create density histograms for each sample size and print the data (number of fakes and not fakes) for each sample size. The density histograms show the distribution of fake and not fake events for each sample size.

#### b) Explain the findings for all three samples

1. Sample Size = 500:


*   Number of Fakes: Varies with each simulation, but it's approximately around 250 since the probability of being fake is 0.5.

*   Number of Not Fakes: Also varies with each simulation, but it's approximately around 250.
*   Findings: In this smaller sample size, the number of fake and not fake events is roughly balanced, as expected from a binomial distribution with a probability of success (being fake) of 0.5.
2. Sample Size = 5000:



*   Number of Fakes: Similar to the 500 sample size, it varies but tends to be close to 2500.

*   Number of Not Fakes: Also similar to the 500 sample size, it varies but tends to be close to 2500.
*   Findings: With a larger sample size, the distribution becomes more consistent, and the number of fake and not fake events is still roughly balanced.

3. Sample Size = 500,000:



*   Number of Fakes: This larger sample size results in a more stable distribution. The number of fake events tends to be around 250,000, half of the total sample size, reflecting the 0.5 probability of being fake.

*   Number of Not Fakes: Similarly, the number of not fake events also tends to be around 250,000.
*   Findings: In this case, with a very large sample size, the distribution closely follows the theoretical binomial distribution with a probability of 0.5. It approaches a nearly perfect 50/50 split between fake and not fake events.


In summary, as the sample size increases, the simulation results tend to approach the expected binomial distribution with a probability of 0.5. The larger the sample size, the closer the observed distribution aligns with the expected distribution, which demonstrates the concept of the law of large numbers, where larger samples tend to produce results closer to the theoretical probabilities.
